1. What is a Decision Tree, and how does it work in the context of classification?



   ans : - A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It represents decisions and their possible outcomes in a tree-like structure. In classification problems, the goal of a Decision Tree is to assign data points to predefined classes.

   The tree consists of three main components: the root node, internal nodes, and leaf nodes. The root node represents the entire dataset, internal nodes represent decision rules based on features, and leaf nodes represent the final class labels. The algorithm works by repeatedly splitting the dataset into smaller subsets based on the feature that best separates the classes. This process continues until a stopping criterion is met, such as reaching a maximum depth or having all records in a node belong to the same class.

2. Explain the concepts of Gini Impurity and Entropy as impurity measures. How do they impact the splits in a Decision Tree?



   ans : - Gini Impurity and Entropy are measures used to evaluate how mixed the classes are within a node of a Decision Tree. They help determine the best feature and threshold for splitting the data.

   Gini Impurity measures the probability of incorrectly classifying a randomly chosen sample if it were labeled according to the class distribution in the node. A lower Gini value indicates a purer node.

3. What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each.



   ans : - Pruning is a technique used to prevent overfitting in Decision Trees.

    Pre-Pruning (Early Stopping) stops the growth of the tree before it becomes too complex. Conditions such as maximum tree depth, minimum samples per node, or minimum information gain are used to stop splitting.

    Advantage of Pre-Pruning: It reduces training time and computational cost because unnecessary branches are not created.

    Post-Pruning (Backward Pruning) allows the tree to grow fully and then removes branches that do not significantly improve performance on validation data.

    Advantage of Post-Pruning: It often produces a more accurate and generalized model because decisions about pruning are made after observing the complete tree structure.

    The main difference is that Pre-Pruning prevents the tree from growing too large, whereas Post-Pruning simplifies a fully grown tree afterward.

4. What is Information Gain in Decision Trees, and why is it important for choosing the best split?



ans : - Information Gain is a metric used to measure how much uncertainty is reduced after splitting a dataset on a particular feature. It is based on the concept of Entropy and helps determine the most informative feature for a split.

5. What are some common real-world applications of Decision Trees, and what are their main advantages and limitations?

Answer:

Decision Trees are widely used in many real-world applications because they are simple and easy to interpret.

Applications of Decision Trees:

a. Medical diagnosis for disease prediction.        
b. Credit risk assessment in banking.

c. Fraud detection in financial transactions.

d.Customer churn prediction in telecommunications.

    Advantages of Decision Trees:

a. Easy to understand and visualize.

b. Requires little data preprocessing.
Can handle both numerical and categorical data.

c. Works for both classification and regression tasks.


    Limitations of Decision Trees:

a. Prone to overfitting if the tree becomes too deep.

b. Small changes in data can result in different tree structures.

c. Can be less accurate than ensemble methods such as Random Forests.
